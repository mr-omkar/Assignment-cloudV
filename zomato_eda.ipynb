```python
currency_rate = {  'Botswana Pula(P)':611,
    'Brazilian Real(R$)':16.66,
     'Dollar($)':83.08,
       'Emirati Diram(AED)':22.62,
        'Indian Rupees(Rs.)':1,
       'Indonesian Rupiah(IDR)':0.0054,
        'NewZealand($)':49.36,
         'Pounds(��)':105.98,
       'Qatari Rial(QR)':22.82,
        'Rand(R)':4.37,
         'Sri Lankan Rupee(LKR)':0.26,
       'Turkish Lira(TL)':3.07
}
```

```python
print(currency_rate)
```

```python
currency_rate.get("Dollar($)")
```

```python
df["converted price"] = df.apply(lambda col:col["Average Cost for two"]* currency_rate.get(col["Currency"]),axis=1)
```

```python
df["converted price"]
```

```python
expensive_hotels = df.sort_values(by="converted price",ascending=False)
```
